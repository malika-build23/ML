```python

import pandas as pd
import requests
```

```python
response = requests.get('https://openlibrary.org/search.json?q=harry+potter')
```

```python
df = pd.DataFrame(response.json()['docs'])[['title','author_name','edition_count' ,'ebook_access','first_publish_year']]
```
