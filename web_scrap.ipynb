```python
company = soup.find_all('div',class_= 'companyCardWrapper')
len(company)
```

```python
name =[]
rating= []
Reviews = []
Salaries = []
Interviews=[]
Jobs =[]
Benefits=[]
Highly_Rated_for = []
Critically_Rated_for = []
for i in company:
    name.append((i.find('h2').text.strip()))
    rating.append(i.find('div', class_='rating_text rating_text--md').text.strip())
    Reviews.append(i.find_all('span','companyCardWrapper__ActionCount')[0].text.strip())
    Salaries.append(i.find_all('span','companyCardWrapper__ActionCount')[1].text.strip())
    Interviews.append(i.find_all('span','companyCardWrapper__ActionCount')[2].text.strip())
    Jobs.append(i.find_all('span','companyCardWrapper__ActionCount')[3].text.strip())
    Benefits.append(i.find_all('span','companyCardWrapper__ActionCount')[4].text.strip())
    try:
        Highly_Rated_for.append(i.find_all('span', class_="companyCardWrapper__ratingValues")[0].text.strip())
    except:
        Highly_Rated_for.append(np.nan) 
    try:
        Critically_Rated_for.append(i.find_all('span', class_="companyCardWrapper__ratingValues")[1].text.strip())
    except:
        Critically_Rated_for.append(np.nan)


d = {'name':name ,'Rating':rating,'Reviews' :Reviews ,'Salaries':Salaries,'Interviews':Interviews,'Jobs':Jobs,'Benefits':Benefits,'Highly_Rated_for':Highly_Rated_for ,'Critically_Rated_for':Critically_Rated_for}

df = pd.DataFrame(d)
df
```

```python
final = pd.DataFrame()
for j in range(1,7):
    
    headers={'User-Agent':'Mozilla/5.0 (Windows NT 6.3; Win 64 ; x64) Apple WeKit /537.36(KHTML , like Gecko) Chrome/80.0.3987.162 Safari/537.36'}
    webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=homepage_explore_companies_widget&page={}'.format(j),headers=headers).text
    soup = BeautifulSoup(webpage,'lxml')
    company = soup.find_all('div',class_= 'companyCardWrapper')
    name =[]
    rating= []
    Reviews = []
    Salaries = []
    Interviews=[]
    Jobs =[]
    Benefits=[]
    Highly_Rated_for = []
    Critically_Rated_for = []
    for i in company:
        name.append((i.find('h2').text.strip()))
        rating.append(i.find('div', class_='rating_text rating_text--md').text.strip())
        Reviews.append(i.find_all('span','companyCardWrapper__ActionCount')[0].text.strip())
        Salaries.append(i.find_all('span','companyCardWrapper__ActionCount')[1].text.strip())
        Interviews.append(i.find_all('span','companyCardWrapper__ActionCount')[2].text.strip())
        Jobs.append(i.find_all('span','companyCardWrapper__ActionCount')[3].text.strip())
        Benefits.append(i.find_all('span','companyCardWrapper__ActionCount')[4].text.strip())
        try:
            Highly_Rated_for.append(i.find_all('span', class_="companyCardWrapper__ratingValues")[0].text.strip())
        except:
            Highly_Rated_for.append(np.nan) 
        try:
            Critically_Rated_for.append(i.find_all('span', class_="companyCardWrapper__ratingValues")[1].text.strip())
        except:
            Critically_Rated_for.append(np.nan)


    d = {'Name':name ,'Rating':rating,'Reviews' :Reviews ,'Salaries':Salaries,'Interviews':Interviews,'Jobs':Jobs,'Benefits':Benefits,'Highly_Rated_for':Highly_Rated_for ,'Critically_Rated_for':Critically_Rated_for}

    df = pd.DataFrame(d)
    
    final = pd.concat([final, df], ignore_index=True)


final




```
